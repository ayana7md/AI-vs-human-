Now, let's read that file into Python. Your CSV file will be read as something called a DataFrame - just a fancy name for a spreadsheet. The command can be broken down as follows:

```python
display(data) #"display" just shows a fancier table
```

There is likely lots and lots of useless data in there. Let's get rid of it! What you will certainly find useless at this point are the first 18 columns - they contain lots of metadata that Qualtrics gathers by default. Check your file - if you see columns like "IPAddress", "LocationLongitude", that means you have that data.

We also have some rows we don't need, namely first and third row. They contain the question numbers, and some tags. Keep in mind, though, that in Python (and many other programming languages) we start counting from 0, not 1. So row 1 becomes row 0, and row 3 becomes row 2.

Drop it, and then check the results.

```python
data.drop(['Start Date',
           'End Date',
           'Response Type',
           'IP Address',
           'Progress',
           'Duration (in seconds)',
           'Finished',
           'Recorded Date',
           'Recipient Last Name',
           'Recipient First Name',
           'Recipient Email',
           'External Data Reference',
           'Location Latitude',
           'Location Longitude',
           'Distribution Channel',
           'User Language',
           'Q_RecaptchaScore',
           'FillerScore',
           'Score'], axis = 1, inplace = True, errors = 'ignore') #dropping the columns which contain stuff that's irrelevant to us
data.drop(index=data.index[0], inplace=True) #dropping the rows which are irrelevant, because Qualtrics just does that
display(data)#see the results of our work
```

```python
cols=[1]
data.drop(data.columns[cols], axis = 1, inplace=True)
display(data)
```

```python
print(data.columns.tolist())
```

```python
data.drop(['AI filler 6 - Click to write Choice 1',
           'human filler 6 - Click to write Choice 1'], axis = 1, inplace = True)
display(data)
```

Now that we have only what we need, it is time to change the spreadsheet into something usable by R. In this example, I have also maintained the metadata ('age_part','gender_part','educ_part').

```python
data_br = pd.melt(data, id_vars=['Response ID'], var_name='Question ID', value_name='Answer')
display(data_br)
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

Since for some of you, not all the questions are displayed to all the participants, it makes sense to remove the questions where the answer was not given, which Qualtrics shows as NaN. That is quite easy to do.

```python
data_br = data_br.dropna(subset = ['Question ID', 'Answer'])
display(data_br)
```

```python
import numpy as np
data_br['is_AI'] = np.where(data_br['Question ID'].str.contains('AI', na=False), 1, 0)
display(data_br)
```

And we're good! The only thing left is to download that spreadsheet. Let's do that. Give it a name, and export the table you've been working on. Then click on 3 dots next to that file (left-hand pane) and choose "Download". That's it!

```python
resps = {element: index for index, element in enumerate (data_br['Response ID'].unique())}
data_br ['Response ID']=data_br ['Response ID'].map(resps)
```
